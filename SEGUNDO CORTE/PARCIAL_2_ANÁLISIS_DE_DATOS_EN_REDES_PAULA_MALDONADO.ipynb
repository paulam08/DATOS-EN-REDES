P2. RFM preliminar

```python
df = pd.read_csv('clientes_segmentos.csv')

print("Columnas disponibles:", df.columns.tolist())

df['R'] = df['dias_desde_ultima']
df['F'] = df['compras']
df['M'] = df['promedio_ticket'] * df['compras']

rfm = df.groupby('id_cliente').agg({
    'dias_desde_ultima': 'sum',
    'compras': 'sum',
    'total_gastado': 'sum'
})

rfm.rename(
    columns={
        'dias_desde_ultima': 'Recency',
        'compras': 'Frequency',
        'total_gastado': 'Monetary'
    },
    inplace=True
)

print(rfm.head())


plt.figure(figsize=(8, 6))
corr_matrix = df[['R', 'F', 'M', 'recomendaciones']].corr()
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            vmin=-1,
            vmax=1,
            fmt=".2f",
            linewidths=0.5)
plt.title('Correlación RFM + Recomendaciones', pad=20)
plt.show()

print("\nAnálisis de correlaciones:")
print("- Frecuencia (F) y Monetary (M):", corr_matrix.loc['F', 'M'].round(2))
print("- Recency (R) y Frequency (F):", corr_matrix.loc['R', 'F'].round(2))
print("- Recomendaciones y Frequency:", corr_matrix.loc['recomendaciones', 'F'].round(2))

df['valor_futuro'] = (0.4 * df['F']) + (0.6 * df['M'])
```

La combinación Frequency (F) + Monetary (M) es clave para predecir el valor futuro del cliente, ya que muestra la mayor correlación (0.65) y captura tanto los hábitos de compra (F) como el impacto económico (M). Además, F correlaciona mejor con las recomendaciones (0.45), mientras que Recency (R) tiene una relación negativa. La fórmula Valor Futuro = 0.6*M + 0.4*F prioriza el gasto histórico sin descuidar la fidelización.

P3. CAC detallado**

```python
import pandas as pd
import matplotlib.pyplot as plt

rfm.head()

cac_data = df.groupby('plataforma').agg(
    total_costo=('costo_adquisicion', 'sum'),
    total_clientes=('id_cliente', 'count')
).reset_index()

cac_data['CAC'] = cac_data['total_costo'] / cac_data['total_clientes']

print(cac_data)

plt.figure(figsize=(8, 5))
plt.bar(cac_data['plataforma'], cac_data['CAC'], color=['#FF6B6B', '#4ECDC4', '#45B7D1'])

plt.title('CAC por Plataforma', fontsize=14)
plt.xlabel('Plataforma')
plt.ylabel('Costo por Cliente (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.3)

for i, cac in enumerate(cac_data['CAC']):
    plt.text(i, cac, f"${cac:.0f}", ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

P4. LTV proyectado

```python
df['tasa_retencion'] = np.where(df['dias_desde_ultima'] <= 90, 0.8,
                               np.where(df['dias_desde_ultima'] <= 180, 0.6, 0.3))

df['LTV'] = df['promedio_ticket'] * (df['compras']/12) * 12 * df['tasa_retencion']


plt.figure(figsize=(10, 5))
sns.histplot(df['LTV'], kde=True, bins=20, color='#4ECDC4')
plt.title('Distribución de LTV a 12 Meses', fontsize=14)
plt.xlabel('LTV (USD)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.show()

top_clientes = df.nlargest(int(len(df)*0.1), 'LTV')[['id_cliente', 'LTV']]
print("Clientes Top 10% por LTV:\n", top_clientes.head())
```

P5. Margen neto


```python
df['margen'] = df['LTV'] - df['costo_adquisicion']

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='plataforma', y='margen', palette=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Distribución de Margen Neto por Plataforma', fontsize=14)
plt.xlabel('Plataforma')
plt.ylabel('Margen Neto (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
margen_por_plataforma = df.groupby('plataforma')['margen'].median().sort_values(ascending=False)
print("\nMargen neto mediano por plataforma:")
print(margen_por_plataforma)

plataforma_optima = margen_por_plataforma.idxmax()
print(f"\nPlataforma con mejor rentabilidad: {plataforma_optima} (Margen mediano = ${margen_por_plataforma.max():.0f})")
```

P6

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = df[['R', 'F', 'M', 'recomendaciones']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette='viridis', s=80)
plt.title('Segmentación de Clientes (K-Means k=3)', fontsize=14)
plt.xlabel('Componente PCA 1')
plt.ylabel('Componente PCA 2')
plt.legend(title='Cluster')
plt.grid(alpha=0.2)
plt.show()

cluster_profile = df.groupby('cluster').agg({
    'R': 'median',
    'F': 'median',
    'M': 'median',
    'recomendaciones': 'mean',
    'id_cliente': 'count'
}).rename(columns={'id_cliente': 'count'})

print("\nPerfil de clusters:")
print(cluster_profile)
```

```python
cluster_names = {
    0: {'name': 'Inactivos', 'slogan': '¡Regresa, te esperamos!'},
    1: {'name': 'Premium', 'slogan': 'Tu lealtad te premia'},
    2: {'name': 'Leales', 'slogan': 'Siempre a tu lado'}
}

df['segmento'] = df['cluster'].map(lambda x: cluster_names[x]['name'])
df['eslogan'] = df['cluster'].map(lambda x: cluster_names[x]['slogan'])

plt.figure(figsize=(12, 7))
scatter = sns.scatterplot(
    x=X_pca[:, 0],
    y=X_pca[:, 1],
    hue=df['segmento'],
    palette=['#FF9E44', '#44C5FF', '#A344FF'],
    s=90,
    style=df['segmento'],
    markers=['o', 's', 'D']
)

for cluster_num in range(3):
    centroid = X_pca[df['cluster'] == cluster_num].mean(axis=0)
    plt.text(
        centroid[0],
        centroid[1],
        cluster_names[cluster_num]['slogan'],
        fontsize=10,
        ha='center',
        va='center',
        bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', boxstyle='round,pad=0.3')
    )

plt.title('Segmentación de Clientes YumFit', fontsize=16, pad=20)
plt.xlabel('Componente Principal 1 (RFM)', fontsize=12)
plt.ylabel('Componente Principal 2 (Recomendaciones)', fontsize=12)
plt.legend(title='Segmento', bbox_to_anchor=(1.05, 1))
plt.grid(alpha=0.1)

print("\nResumen de segmentos:")
segment_summary = df.groupby('segmento').agg({
    'R': 'median',
    'F': 'median',
    'M': 'median',
    'recomendaciones': 'mean',
    'id_cliente': 'count'
}).rename(columns={'id_cliente': 'clientes'})

print(segment_summary)
```

```python
percentil_75 = df['margen'].quantile(0.75)
df['high_margin'] = (df['margen'] >= percentil_75).astype(int)

print("\nDistribución de High Margin:")
print(df['high_margin'].value_counts(normalize=True).mul(100).round(1))

plt.figure(figsize=(8, 5))
sns.barplot(x=df['high_margin'].value_counts().index,
            y=df['high_margin'].value_counts(),
            palette=['#FF6B6B', '#4ECDC4'])
plt.title('Clientes High Margin (Percentil 75)', fontsize=14)
plt.xlabel('High Margin (1=Sí, 0=No)')
plt.ylabel('Número de Clientes')
plt.xticks([0, 1], ['No', 'Sí'])
plt.grid(axis='y', linestyle='--', alpha=0.3)

for i, count in enumerate(df['high_margin'].value_counts()):
    plt.text(i, count, f"{count}\n({df['high_margin'].value_counts(normalize=True).iloc[i]*100:.1f}%)",
             ha='center', va='bottom', fontsize=10)

plt.show()

print("\nCaracterísticas High Margin vs otros:")
high_margin_profile = df.groupby('high_margin').agg({
    'R': 'median',
    'F': 'median',
    'M': 'median',
    'plataforma': lambda x: x.mode()[0],
    'segmento': lambda x: x.mode()[0]
})

print(high_margin_profile)
```

P9

```python
cluster_metrics = df.groupby('cluster').agg({
    'margen': 'mean',
    'id_cliente': 'count'
}).rename(columns={'id_cliente': 'n_clientes'})

cluster_metrics['margen_total'] = cluster_metrics['margen'] * cluster_metrics['n_clientes']

presupuesto = 10000

cluster_optimo = cluster_metrics['margen'].idxmax()
inversion_esc1 = {cluster: presupuesto if cluster == cluster_optimo else 0 for cluster in range(3)}

top2_clusters = cluster_metrics['margen'].nlargest(2).index
inversion_esc2 = {cluster: presupuesto/2 if cluster in top2_clusters else 0 for cluster in range(3)}

margen_total = cluster_metrics['margen_total'].sum()
inversion_esc3 = {cluster: (presupuesto * cluster_metrics.loc[cluster, 'margen_total'] / margen_total)
                  for cluster in range(3)}

def calcular_roi(inversion_dict):
    return sum([inversion_dict[cluster] * cluster_metrics.loc[cluster, 'margen'] for cluster in range(3)])

roi_esc1 = calcular_roi(inversion_esc1)
roi_esc2 = calcular_roi(inversion_esc2)
roi_esc3 = calcular_roi(inversion_esc3)

resultados = pd.DataFrame({
    'Escenario': ['1: Enfoque en 1 cluster', '2: Dos clusters clave', '3: Proporcional al margen'],
    'Inversión': [inversion_esc1, inversion_esc2, inversion_esc3],
    'ROI_esperado': [roi_esc1, roi_esc2, roi_esc3]
})

print("\nComparación de escenarios:")
print(resultados[['Escenario', 'ROI_esperado']])

plt.figure(figsize=(10, 5))
plt.barh(resultados['Escenario'], resultados['ROI_esperado'],
        color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('ROI Esperado por Escenario ($10,000 USD)', fontsize=14)
plt.xlabel('ROI Esperado (USD)')
plt.grid(axis='x', linestyle='--', alpha=0.3)

for i, roi in enumerate(resultados['ROI_esperado']):
    plt.text(roi, i, f"${roi:,.0f}", ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()

mejor_escenario = resultados.loc[resultados['ROI_esperado'].idxmax(), 'Escenario']
print(f"\nRecomendación: Optar por '{mejor_escenario}' con ROI esperado = ${max(roi_esc1, roi_esc2, roi_esc3):,.0f}")
```

El dataset solo tiene clientes que siguen activos, o sea que los que se fueron antes no aparecen. Esto puede hacer que el LTV parezca más alto de lo real.

Las recomendaciones vienen solo de los que ya participan en referidos, así que no reflejan a todos los clientes por igual. Los que no comparten no están en los datos.

Agregar datos de clientes que ya no están para comparar y ver por qué se fueron.

Cruzar con info demográfica (edad, ciudad) para ver si faltan segmentos importantes.
